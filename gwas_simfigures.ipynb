## Simulated GWAS PLOTS: QQ & MANHATTAN

```r
files <- list.files("/tscc/projects/ps-gaultonlab/mpittigher/random/GWAS_results", pattern = "*.mlma", full.names = TRUE)

for (f in files) {
  
  sumstats <- read.table(f, header = TRUE, sep = "\t")
  
  colnames(sumstats)[colnames(sumstats) == "Chr"] <- "CHR"
  colnames(sumstats)[colnames(sumstats) == "bp"] <- "BP"
  colnames(sumstats)[colnames(sumstats) == "p"] <- "P"
  
  sumstats <- sumstats[!is.na(sumstats$P) & sumstats$P > 0, ]
  
  trait_name <- tools::file_path_sans_ext(basename(f))
  
  manhattan(sumstats,
            chr = "CHR",
            bp = "BP",
            snp = "SNP",
            p = "P",
            main = paste("Simulated GWAS:", trait_name))

    qq(sumstats$P)
}
```

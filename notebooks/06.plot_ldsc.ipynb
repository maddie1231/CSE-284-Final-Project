```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
ldsc_df = pd.read_csv("../ldsc/ldsc.all.sumstat.results.txt", sep = "\t")
ldsc_df = ldsc_df.sort_values("h2 - LDSC")
ldsc_df.head()
```

```python
plt.figure(figsize=(6,7))
colors = plt.cm.Set2(np.linspace(0, 1, len(ldsc_df)))

for i, row in ldsc_df.iterrows():
    plt.errorbar(
        i,
        row["h2 - LDSC"],
        yerr=row["se - LDSC"],
        fmt='o',
        capsize=3,
        color=colors[i]
    )


plt.xticks(range(len(ldsc_df)), ldsc_df["Trait"], rotation=45, ha="right")

plt.ylabel("Heritability (h2)")
plt.xlabel("Trait")

plt.tight_layout()
plt.show()
```
